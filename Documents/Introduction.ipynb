# 2. Nhập dữ liệu

## 2.1 Nhập số liệu trực tiếp: c()
Ví dụ: Ta muốn có số liệu về độ tuổi và insulin của 1 bệnh nhân như sau, và muốn nhập vào R:

age = c(50, 62, 60, 40, 48, 47, 57, 70, 48, 67)

insulin = c(16.5, 10.8, 32.3, 19.3, 14.2, 11.3, 15.5, 15.8, 16.2, 11.2)

Mỗi biến age, insulin được xem là 1 đối tượng, data.frame cũng là 1 đối tượng. data.frame có thể hiểu đơn giản là 1 bảng 2 chiều, với chiệu cột là các đặc trưng và chiều hàng là mỗi đối tượng (như bệnh nhân). Vì vậy ta cần phải nhấp hai đối tượng này thành một data.frame. 

Ví dụ: benhnhan = data.frame(age, insulin)

Để kiểm tra xem trong benhnhan có gì chúng ta chỉ cần gõ >benhnhan, và R sẽ báo cáo:
![alt](Hinh_3.png)

Nếu chúng ta cần lưu dữ liệu ở đường dẫn nào đó, dùng hàm save(benhnhan, file="benhnhan.rda"). Lưu ý: nhớ hãy dùng lệnh setw() để cấu hình đường dẫn, nơi mà dữ liệu sẽ được đặt. Và file benhnhan.rda sẽ xuất hiện đích của đường dẫn đó.

Sử dụng edit(data.frame()). R cung cấp cho chúng ta một cửa sổ mới, nhằm để điền dữ liệu.

## 2.3 Nhập số liệu từ một text file: read.table
Ta đọc bảng (dữ liệu đã được tạo sẵn). 

Trước tiên, sử dụng setwd() để cấu hình đường dẫn chứa dữ liệu. 

Dùng hàm chol = read.table("file.txt", header=TRUE). header = TRUE nghĩa là đọc dòng đầu tiên của bảng. Kiểm tra xem R đã đọc hết các dữ liệu hay chưa
bằng lệnh >chol hay >names(chol) (names cho biết tên các cột ở trong chol). 

## 2.4 Nhập số liệu từ Excel: read.csv
Cần đưa file excel đó về định dạng csv. Vào Excel, chọn File -> Save as, chọn Save as type "CSV (Comma delimited)". 

Sau khi có file csv. Ta cấu hình đường dẫn bằng setwd(). Tiếp theo dùng read.csv như read.table.

## 2.5 Nhập số liệu từ một SPSS: read.spss
Phần mềm SPSS lưu dữ liệu dưới dạng "sav". Do đó, ta cần dùng lệnh read.spss trong package có tên là foreign.

Đầu tiên, dùng lệnh library(foreign) để truy nhập foreign.

Thứ hai, là lệnh read.spss:
setwd()

testo = read.spss("testo.sav", to.data.frame=TRUE). Đọc số liệu từ testo.sav và cho vào một data.frame có tên là testo.


## 2.6 Thông tin cơ bản về dữ liệu
Sau khi đã đưa dữ liệu vào 1 data.frame. Để tìm hiểu xem trong dữ liệu này có gì, chúng ta có thể nhập R vào như sau:

Dẫn cho R biết chúng ta muốn xử lý chol (data.frame) bằng cách dùng lệnh attach(arg) với arg là tên dữ liệu. Chả hiểu ý của sách nói về hàm này là làm gì.

is.data.frame(arg) - kiểm tra xem biến arg có phải là data.frame hay không.

dim(arg) - kết quả cho ra số lượng cột và số lượng hàng. 

names(arg) - tên của từng cột theo thứ tự trong bảng.

table(tên 1 cột của arg) - nhớ phải dùng attach(arg) trước - kết quả là số lượng của mỗi giá trị trong đó. Ví dụ trong cột đó có 3 giá trị nam, Nam, Nu,
    thì kết quả cho ra là bảng 2 hàng, 3 cột, hàng đầu tiên lần lượt là nam, Nam, Nu, hàng thứ 2 là số lần xuát hiện của các giá trị đó.


# 3. Biên tập dữ liệu
Cách tổ chức dữ liệu sao cho R có thể phân tích một cách hữu hiệu. Nhiều khi trong phân tích thống kê, chúng ta cần phải tập trung số liệu thành một nhóm, hay tách rời thành từng nhóm, hay thay thế từ kí tự (characters) sang số (numeric) cho tiện việc tính toán. Trước khi biên tập, phải hoàn thành các code sau:

setwd("c:/works/stats")

chol = read.table("chol.txt", header=TRUE)

attach(chol)

## 3.1 Kiểm tra số liệu trống không (missing value)
Có một số thống kê cần đầy đủ số liệu mới tính toán được. Khi đó, ta sẽ cần xóa các mẫu có ô chứa số liệu trống. 

chol.new = na.omit(chol) - biến chol.new sẽ là data.frame mới chứa dữ liệu lấy từ chol của các mẫu có dữ liệu trống không (na).
